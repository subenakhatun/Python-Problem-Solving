Stack (LIFO principle) - Last In, First Out
আমি এই মাসে নতুন বাসায় উঠেছি। বাসায় উঠার সময় আমি আমার জিনিসপত্র একটার উপর একটা করে একটা বড় ব্যাগ (container) এর ভিতর রেখেছিলাম।
যখন আমি জিনিসগুলো রাখছিলাম:
•	প্রথমে আমি যেটা রেখেছি, সেটা একদম নিচে গেছে 
•	তারপর আরেকটা, তার উপর গেছে 
•	এভাবে শেষ পর্যন্ত সব জিনিস রাখা হয়েছে 
 ফলে যেটা সবশেষে রেখেছিলাম, সেটা একদম উপরে ছিল।


 
বাসায় এসে যখন আমি জিনিসগুলো বের করতে শুরু করলাম:
•	প্রথমে আমি উপরের জিনিসটা বের করলাম (যেটা শেষ রেখেছিলাম) 
•	তারপর তার নিচেরটা 
•	এভাবে একে একে সব বের করলাম 
 ফলে দেখা গেল:
•	যেটা প্রথমে রেখেছিলাম সেটা একদম শেষে বের হয়েছে 
•	আর যেটা শেষে রেখেছিলাম সেটা আগে বের হয়েছে
