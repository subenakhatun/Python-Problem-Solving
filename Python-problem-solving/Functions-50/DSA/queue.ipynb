ধরা যাক, একটি কোচিং সেন্টারে মাসের শুরুতে অভিভাবকরা তাদের সন্তানদের বেতন পরিশোধ করতে আসেন। তারা একসাথে এলেও, সবাই একটি লাইনে দাঁড়িয়ে ধীরে ধীরে সামনে এগোন।

যিনি আগে আসেন, তিনি লাইনের প্রথমে দাঁড়ান। এরপর একে একে নতুন অভিভাবকরা লাইনের শেষে যোগ দেন। বেতন দেওয়ার সময়ও একই নিয়ম অনুসরণ করা হয়—যিনি সবার আগে এসেছেন, তিনি প্রথমে বেতন পরিশোধ করেন এবং কাজ শেষ হলে লাইন থেকে বের হয়ে যান। তারপর পরবর্তী ব্যক্তি একইভাবে বেতন দেন।

এই পুরো প্রক্রিয়াটি “Queue (কিউ)” ডাটা স্ট্রাকচারের একটি বাস্তব উদাহরণ। এখানে “First In, First Out (FIFO)” নিয়মটি অনুসরণ করা হয়, অর্থাৎ যে আগে লাইনে ঢোকে, সে-ই আগে তার কাজ শেষ করার সুযোগ পায়।

```python
from collections import deque
q = deque()
for i in range(1,5):
    ask_book = input(" ")
    q.append(ask_book)
print(f"item-empty:{q}")
after_remove = q.popleft()
print(f"item-empty:{after_remove}")
```
