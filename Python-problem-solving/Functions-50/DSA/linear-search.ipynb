## Linear Search
কোচিং সেন্টারের বুকশেলফে ৩য় শ্রেণি থেকে ১০ম শ্রেণি পর্যন্ত সব বিষয়ের বই সাজানো আছে। এখন ইসরাতের একটি নির্দিষ্ট বই প্রয়োজন—“Chemistry ২০২৬”।
ইসরাত বুকশেলফে গিয়ে প্রথম বইটি থেকে শুরু করে একে একে প্রতিটি বইয়ের নাম দেখতে থাকে এবং তার প্রয়োজনীয় বইয়ের সাথে মিলিয়ে দেখে। এভাবে ধারাবাহিকভাবে (একটার পর একটা) খুঁজতে খুঁজতে এক পর্যায়ে সে তার কাঙ্ক্ষিত বইটি পেয়ে যায়।
এইভাবে তালিকার শুরু থেকে শেষ পর্যন্ত একে একে খোঁজার পদ্ধতিকেই Linear Search (লিনিয়ার সার্চ) বলা হয়।

```python
booklist = ['bangla','english','math','chemistry','physics','higher math','isalm','ict']
searcBook = 'chemistry'
#n = len(booklist)
#print(n)
for i in range(len(booklist)):
    if booklist[i] == searcBook:
        print(i)
```
