## Binary Search
কোচিং সেন্টারে শিক্ষার্থীদের সাপ্তাহিক পরীক্ষার নম্বর একটি তালিকায় সাজানো আছে। তালিকাটি এমনভাবে সাজানো যে, যারা বেশি নম্বর পেয়েছে তারা শুরুতে এবং কম নম্বর পাওয়া শিক্ষার্থীরা শেষে রয়েছে (অর্থাৎ descending order-এ sorted)।<br>

এখন ম্যানেজমেন্ট থেকে বলা হলো—<br>
“Class 8-এর যে শিক্ষার্থীরা ৮৯ পেয়েছে, তাদের খুঁজে বের করো।”<br>

তখন দায়িত্বপ্রাপ্ত ব্যক্তি যেভাবে খোঁজ করবে—<br>

প্রথমে তালিকার মাঝখানের শিক্ষার্থীর নম্বর দেখবে <br>
যদি নম্বর ৮৯-এর চেয়ে বেশি হয়, তাহলে বুঝবে ৮৯ পাওয়া শিক্ষার্থীরা ডান পাশে (কম নম্বরের দিকে) আছে<br>
যদি নম্বর কম হয়, তাহলে বাম পাশে (বেশি নম্বরের দিকে) খুঁজবে<br>
এরপর সেই অংশের আবার মাঝখানে গিয়ে একইভাবে তুলনা করবে<br>
এভাবে বারবার তালিকাকে অর্ধেক করে খুঁজতে খুঁজতে একসময় ৮৯ পাওয়া শিক্ষার্থীদের পাওয়া যাবে

```python
markslist = [78,63,21,98,78,53,14,36,52,89]

sortedlist = sorted(markslist)
print(sortedlist)

findmarks = 89
def BinarySearch(sortedlist,findmarks):
    leftstart = 0
    rightend = len(sortedlist) - 1

    while leftstart <= rightend:
        mid = (leftstart + rightend) // 2
        print(f'checking {sortedlist[mid]}')

        if sortedlist[mid] == findmarks:
            return "Found at index:", mid
            break

        elif sortedlist[mid] > findmarks:
            rightend = mid - 1   # ✔ correct

        else:
            leftstart = mid + 1  # ✔ correct

    else:
        return"Not found"
BSaerch = BinarySearch(sortedlist,findmarks)
print(BSaerch)
```
